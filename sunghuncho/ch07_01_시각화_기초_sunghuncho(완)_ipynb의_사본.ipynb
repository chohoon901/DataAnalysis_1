```python
# https://wikidocs.net/137791
def draw_line(style):
  plt.title("나이와 재산의 상관관계")
  plt.plot(
      range(10,50,10),
      [1,4,9,16],
      style
    )
  plt.xlabel('나이')
  plt.ylabel('재산')
  plt.show()
```

```python
draw_line('bo:')
```

### 그리드 설정 (grid)


```python
# 보조선 활성화/비활성화 : plt.grid(T/F)
def draw_line(grid=False,style="b-"):
  plt.title("나이와 재산의 상관관계")
  plt.plot(
      range(10,50,10),
      [1,4,9,16],
      style
    )
  plt.xlabel('나이')
  plt.grid(grid)
  plt.ylabel('재산')
  plt.show()
```

```python
draw_line()
```

```python
draw_line(True)
```

### 여러 개의 선 그어보기

```python
t=np.arange(0,5,0.2)
t
```

```python
# 1. 여러 개의 선 그리기 (하나의 plot 사용)
plt.title("라인 플롯에서 여러개의 선")
plt.plot(
    t,t,"r--",
    t,t**2,"bs:",
    t,t**3,"g^-",
)
plt.show()
```

```python
# 2. 하나의 plot이 아니라 여러 개 -> 겹치는.
plt.title("여러개의 plot 명령으로 한 그림에 겹쳐서 표현")
plt.plot((np.arange(4)+1)**2)
plt.show()
plt.plot((np.arange(4)-1)**3)
plt.show()
```

```python
# 여러 개의 플롯이 겹쳤을 경우 라벨 표시
X=np.linspace(-np.pi,np.pi,256)
X
```

```python
# 코사인, 사인 곡선
C,S=np.cos(X),np.sin(X)
C,S
```

```python
plt.title("삼각함수 plot")
plt.plot(X,C)
plt.plot(X,S)
plt.show()
```

```python
plt.title("삼각함수 plot")
plt.plot(X,C,label="cosine")
plt.plot(X,S,label="sine")
plt.legend(loc=2)
plt.show()
```

#### `subplot`
* plot 명령을 여러 번 실행 => 같은 그래프 내부에 여러 직선 표시
* subplot : 행과 열 단위로 묶어서 여러 데이터를 병행해서 표시

```python
import numpy as np
```

```python
# 위아래 행 배치
# 행, 열, 서브플롯의 위치
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(np.arange(5)**2)
plt.title("차트1")
plt.subplot(2,1,2)
plt.plot(np.arange(10),np.arange(10)**3,"r--")
plt.title("차트2")
plt.tight_layout() # 안겹치게 레이아웃 조정
plt.show()
```

```python
# 양옆 열 배치
# 행, 열, 서브플롯의 위치
plt.subplot(1,2,1)
plt.plot(np.arange(5)**2)
plt.title("차트1")
plt.subplot(1,2,2)
plt.plot(np.arange(10),np.arange(10)**3,"r--")
plt.title("차트2")
plt.tight_layout() # 안겹치게 레이아웃 조정
plt.show()
```

```python
# 2x2 플롯 배치
plt.subplot(2,2,1)
plt.plot(np.arange(5)**2)
plt.title("차트1")
plt.xlabel("x축")
plt.ylabel("y축")
plt.subplot(2,2,2)
plt.plot(np.arange(10),np.arange(10)**3,"r--")
plt.title("차트2")
plt.subplot(2,2,3)
plt.plot(np.arange(5)**2)
plt.title("차트3")
plt.subplot(2,2,4)
plt.plot(np.arange(10),np.arange(10)**3,"r--")
plt.title("차트4")
plt.tight_layout() # 안겹치게 레이아웃 조정
plt.show()
```

```python
y=[2,3,1]
x=np.arange(3)
xlabel=["가","나","다"]
# plt.xticks(x,xlabel) 
plt.bar(x,y)
plt.show()
```

```python
y=[2,3,1]
x=np.arange(3)
xlabel=["가","나","다"]
plt.xticks(x,xlabel) 
plt.bar(x,y)
plt.show()
```

```python
y=[2,3,1]
x=np.arange(3)
xlabel=["가","나","다"]
plt.title("바 차트")
plt.xticks(x,xlabel) 
plt.yticks(sorted(y)) # sticks : 값에 이름 붙여줌 
plt.bar(x,y)
plt.show()
```

```python
y=[2,3,1]
x=np.arange(3)
xlabel=["가","나","다"]
plt.title("가로바 차트")
plt.yticks(x,xlabel) 
plt.xticks(sorted(y)) # sticks : 값에 이름 붙여줌  # 차트 회전시 x와 y축 변경해야함
# plt.bar(x,y)
plt.barh(x,y)
plt.show()
```

```python
people=["James","John","Kate","Alice"]
y_pos=np.arange(4)
performance=3+10*np.random.rand(4)
error=np.random.rand(4)

plt.title("Bar-h Chart")
plt.grid(True)
plt.barh(y_pos,performance,xerr=error,alpha=0.4)
plt.yticks(y_pos,people)
plt.xlabel("x축")
plt.show()
```

### 플롯 종류 선택 (`kind`)

```python
iris : pd.DataFrame = sns.load_dataset('iris') # 붓꽃
titanic : pd.DataFrame = sns.load_dataset('titanic') # 타이타닉호 데이터
```

```python
iris
```

```python
list(iris.columns)
```

```python
# kind = 종류 => 해당 종류 plot 교체
iris[:5].plot(kind="bar",rot=0,ylim=(3,6))
# plt.title("꽃받침의 길이 시각화")
# plt.xlabel("Data")
# plt.ylabel('꽃받침의 길이')
plt.tight_layout()
plt.show()
```

```python
# kind 를 사용할 필요 없이 plot 다음에 직접적으로 해당 차트의 메소드 호출해서 사용
iris[:5].plot.bar(rot=0)
plt.ylim(0,7)
```

```python
# 파이 차트 : 원형, 비율을 나타낼 때
df2=titanic["class"].value_counts()
df2
```

```python
df2.plot.pie(autopct="%.2f%%")
plt.axis('equal')
plt.show()
```

```python
# 히스토그램
# - 가로축에 분포(값), 세로축에 빈도(몇개...)를 시각화
iris.plot.hist()
```

```python
# 커널 밀도 그래프
# - 부드럽게 그린 히스토그램
iris.plot.kde()
```

```python
# 박스플롯
# 5대 요약 수치 제공 + 이상치 추론
iris.plot.box()
```

```python
iris.boxplot(by='species')
plt.tight_layout(pad=3,h_pad=1)
plt.title("각 Feature의 종 별 데이터에 대한 Box Plot")
plt.show()
```

```python
# 산점도 : 두 데이터의 관계를 점으로 표현
iris.plot.scatter('petal_width', 'petal_length')
```
